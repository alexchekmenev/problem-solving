```python
import numpy as np
import matplotlib.pyplot as plt
import time

n = 12
p = 6
k = 5

# y = [1, 2, 10, 9, 8, 6, 7, 3, 11, 5, 4]
# y = [10, 3, 11, 6, 7, 4, 1, 5, 8, 12, 9, 2] -- True не подотрезок


def gen(pos, step, y, n):
    seq = [y[pos]]
    l = pos - 1
    r = pos + 1
    while len(seq) < step:
        if l >= 0 and r < n:
            if y[l] < y[r]:
                seq.append(y[l])
                l -= 1
            else:
                seq.append(y[r])
                r += 1
        elif l >= 0:
            seq.append(y[l])
            l -= 1
        elif r < n:
            seq.append(y[r])
            r += 1
    return seq

def run():
    np.random.seed(seed=int(time.time()))
    y = list(map(lambda xx: xx + 1, np.random.permutation(n).tolist()))
    print(y)
    x = range(1, n + 1)
    value = y[p - 1]

    ans = 0
    valid_x = []
    valid_y = []
    for i in range(0, len(x)):
        seq2 = gen(i, k, y, n)
        is_valid = seq2[k - 1] == value
        if is_valid:
            valid_x.append(i + 1)
            valid_y.append(y[i])
        ans += is_valid
        # print(i + 1, is_valid, seq2)

    print(f'ans = {ans}')

    plt.plot(x, y, 'bo-')
    plt.plot(valid_x, valid_y, 'ro')
    plt.plot(p, value, 'go')
    plt.xticks(x)
    plt.yticks(x)

    plt.show()
```

```python
run()
```
